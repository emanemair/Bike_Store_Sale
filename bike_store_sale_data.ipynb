```python
# Aggregate by year
df_agg = df.resample('Y', on='Date').sum()

# Plot bar chart
plt.figure(figsize=(10,6))
plt.bar(df_agg.index.year, df_agg['Revenue'], color='green')
plt.title('Yearly Revenue')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.grid(True, axis='y')
plt.show()
```

```python
import seaborn as sns

# Pivot the data to create a heatmap of monthly revenue over years
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
heatmap_data = df.pivot_table(index='Year', columns='Month', values='Revenue', aggfunc='sum')

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".0f")
plt.title('Monthly Revenue Heatmap')
plt.show()
```
